## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

total_pages = df.shape[0]
declining_rate = df["is_declining_label"].mean()
stale_and_visible = ((df["days_since_last_update"] >= 180) & (df["impressions_90d"] >= 500)).sum()

print(f"Total pages in starter dataset: {total_pages}")
print(f"Share currently labeled declining: {declining_rate:.1%}")
print(f"Pages that are both stale (180+ days) and still visible (500+ impressions): {stale_and_visible}")


# Out of 30,000 pages, roughly 54.2% are currently labeled declining, and 17 pages are
# both stale (180+ days since update) and still getting meaningful traffic (500+ impressions) —
# a small but concrete pool of high-confidence candidates worth prioritizing over reviewing at random.
# Additionally, my Week 1 run showed a learned ranking (Precision@50 = 0.740)
# beating a hand-written rule (Precision@50 = 0.240) by ~3x, which is direct
# evidence this lane's core method — learned ranking over rules — has real signal.
```
