## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
My lane (Refresh / Content Opportunity Scoring) is primarily a **classification**
task, combined with **ranking** for the final output. I predict whether a page is
declining (a yes/no classification), then use the model's predicted probability to
rank all pages from highest to lowest priority. Classification alone tells me
which pages look risky; ranking is what actually gets used, since a reviewer
works down a prioritized list rather than treating every "yes" prediction equally.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
My target is `is_declining_label`, defined as `trend_direction == "down"`.
This is a **proxy label**, not a true observed future outcome — it's a bucket
calculated from the current data window, not something that happened after
a defined decision point. A stronger version of this label (planned for later
in the internship) would use a genuine future outcome, e.g. "features from the
prior 90 days -> decline over the next 30 days." For now, this proxy is good
enough to prove the workflow, but I'm treating it as a starting point, not
the final target.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
My metric is **Precision@50** (and I'll also track Precision@20). This matches
the actual business decision: a reviewer has limited time and works through a
fixed-size list, so what matters is "of the top 50 pages flagged, how many are
genuinely worth reviewing?" — not overall accuracy across all 30,000 pages,
which would reward the model for being right about pages nobody's going to look
at anyway. My Week 1 baseline scored 0.240 on this metric; the trained model
scored 0.740 — a "good" number for my model going forward is anything that
clearly beats 0.240 and ideally holds above ~0.5-0.7 depending on data used.
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
df["is_declining_label"] = df["trend_direction"].str.lower().eq("down").astype(int)

features = ["content_age_days", "days_since_last_update", "impressions_90d",
            "avg_position", "ctr", "word_count"]

df[features + ["is_declining_label"]].head(5)

One row = one content page (one URL/piece of content), described by its
observable signals (age, freshness, traffic, position, click rate, length)
at a single point in time, plus the label showing whether it's currently
trending down. This is the exact grain my model will train and rank on.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
A fixed if-statement rule can only combine 2-3 conditions before it becomes
unreadable, and it has to pick fixed thresholds by hand (e.g. "180+ days AND
500+ impressions"). But the real relationship between these signals and
decline is not that clean — different combinations of position, CTR, age,
and traffic interact in ways a human can't fully anticipate by hand. My own
Week 1 result shows this directly: my simple hand rule reached Precision@50
of 0.240, while a random forest — which can weigh many signals together and
find non-obvious combinations — reached 0.740, roughly 3x better. That gap is
the evidence that this pattern is too messy for a fixed rule to capture well.
```
